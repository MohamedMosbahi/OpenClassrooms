### Gestion des données manquantes

```python
df[(df['order_status'] == 'delivered') & (df['product_category_name'].isna())]
```

Nous n'allons pas imputer les valeurs manquantes des catégories de produit car elles ne représentent que 2% environ des données.
Cela nous permet de garder des données relativement fiables avant d'essayer de rassembler les clients selon leurs ressemblances.

```python
df_clean = df.copy()
df_clean = df_clean.dropna(subset=['product_category_name'])
df_clean.isna().sum()
```

Les données manquantes restantes sont liées aux dates.
Nous allons donc imputer ces valeurs en se servant de la durée moyenne entre chaque étape d'une commande.

Pour cela, il faut d'abord calculer ces durées.

```python
import datetime
```

```python
# Colonnes correspondant aux dates et heures des différentes étapes de chaque commande.

heure = ['order_purchase_timestamp',
              'order_approved_at', 
              'order_delivered_carrier_date',
              'order_delivered_customer_date',
              'order_estimated_delivery_date',
              'shipping_limit_date',
              'review_creation_date',
              'review_answer_timestamp'
             ]
```

```python
# Type de ces colonnes

df_clean[heure].dtypes
```

```python
# Conversion au format 'date'.

for col in heure:
    df_clean[col] = pd.to_datetime(df_clean[col], errors='coerce')
    
df_clean = df_clean.reset_index(drop=True)
df_clean
```

```python
# Vérification

df_clean[heure].dtypes                                       
```

```python
# Aperçu des durées entre l'estimation de la livraison et la validation de la commande.

plt.figure(figsize=(20,12))
plt.scatter(x=df_clean.index,
            y=(df_clean['order_estimated_delivery_date'] - df_clean['order_purchase_timestamp'])/(60*60*24))
```

```python
# Création de colonnes correspondant aux durées par rapport à la validation de la commande.


duree = ['duree_order_approved_at',
         'duree_order_delivered_carrier_date',
         'duree_order_delivered_customer_date',
         'duree_order_estimated_delivery_date',
         'duree_shipping_limit_date',
         'duree_review_creation_date',
         'duree_review_answer_timestamp'
        ]

i=1
for col in duree:
    df_clean[col] = df_clean[heure[i]] - df_clean['order_purchase_timestamp']
    i+=1
```

```python
df_clean
```

```python
df_clean.isna().sum()
```

Les colonnes correspondant aux durée ont bien été ajoutées et elles ont les mêmes valeurs manquantes que les dates qui leur ont permis d'être calculées.

```python
# Imputation des valeurs manquantes de durées par la moyenne.

for col in duree:
    df_clean[col].fillna(df_clean[col].mean(), inplace=True)
```

```python
# Fonction qui va imputer les valeurs manquantes des dates de chaque étape.

def remplissage(col, duree_col):
    missing = pd.isna(df_clean[col]).index
    start = df_clean['order_purchase_timestamp']
    duree = df_clean[duree_col]
    df_clean.iloc[missing, col] = start[missing] + duree[missing]
    return df_clean
```

```python
# Application de la fonction aux colonnes 'date' avec des valeurs manquantes.

df_clean['order_approved_at'].fillna(df_clean['order_purchase_timestamp'] + df_clean['duree_order_approved_at'], inplace=True)
df_clean['order_delivered_carrier_date'].fillna(df_clean['order_purchase_timestamp'] + df_clean['duree_order_delivered_carrier_date'], inplace=True)
df_clean['order_delivered_customer_date'].fillna(df_clean['order_purchase_timestamp'] + df_clean['duree_order_delivered_customer_date'], inplace=True)
```

```python
# Vérification du remplissage des valeurs manquantes

df_clean.isna().sum()
```

```python
# Imputation par la moyenne pour les dernières valeurs manquantes.

products = ['product_weight_g',
            'product_length_cm',
            'product_height_cm',
            'product_width_cm'
           ]

for col in products:
    df_clean[col].fillna(df_clean[col].mean(),inplace=True)
```

## KMeans - Baseline

```python
from sklearn.cluster import KMeans
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
# Prétraitement pour les données numériques.
rfm_scaled = df_rfm.copy()


numeric_features = ['Frequency', 'Monetary']
ss = StandardScaler()
rfm_scaled[numeric_features] = ss.fit_transform(rfm_scaled[numeric_features])

# Prétraitement pour les données catégorielles.
categorical_features = ['customer_unique_id', 'Recency']
le = LabelEncoder()

for col in categorical_features:
    rfm_scaled[col] = le.fit_transform(rfm_scaled[col])
```

```python
rfm_scaled
```

```python
from sklearn.metrics import silhouette_score, davies_bouldin_score, adjusted_rand_score
```

```python
# Enregistrement des scores du modèle.

inertie = []
db = []
silhouettes = []

for num_clusters in range(2, 11):

    cls = KMeans(n_clusters=num_clusters, n_init=1, init='random')
    cls.fit(rfm_scaled)

    inertie.append(cls.inertia_)

    clusters = cls.labels_
    db.append(davies_bouldin_score(rfm_scaled, clusters))

    silh = silhouette_score(rfm_scaled, cls.labels_)
    silhouettes.append(silh)
```

# Feature Engineering

```python
pd.set_option('display.max_columns', 45)
df_clean
```

```python
# Aggrégation des données pertinentes par client unique 

df_fe = df_clean.groupby('customer_unique_id').agg({'order_approved_at' : lambda x: (now-x),
                                                    'order_id' : pd.Series.nunique,
                                                    'order_item_id' : np.sum,
                                                    'price' : np.sum,
                                                    'freight_value' : np.sum,
                                                    'review_score' : np.mean, 
                                                    'product_category_name' : pd.Series.nunique,
                                                    'product_weight_g' : np.mean,
                                                    'product_length_cm' : np.mean,
                                                    'product_height_cm' : np.mean,
                                                    'product_width_cm' : np.mean,
                                                    'credit_card' : np.mean
                                                   }).reset_index()

# Nouveaux noms de colonnes

df_fe.columns = ['Customer',
                 'Recency',
                 'Orders',
                 'Quantity',
                 'Price',
                 'Freight',
                 'Score' ,
                 'Categories',
                 'Weight',
                 'Length',
                 'Height',
                 'Width',
                 'Credit_card'
                ]
```

```python
# Création de variables calculées

df_fe['Order_Mean_Price'] = df_fe['Price']/df_fe['Orders']         # Prix par commande
df_fe['Order_Mean_Freight'] = df_fe['Freight']/df_fe['Orders']     # Coût de frêt par commande
df_fe['Order_Mean_Quantity'] = df_fe['Quantity']/df_fe['Orders']   # Nombre de produits par commande

df_fe['Product_Mean_Price'] = df_fe['Price']/df_fe['Quantity']     # Prix moyen pour 1 produit de la même commande
df_fe['Product_Mean_Freight'] = df_fe['Freight']/df_fe['Quantity'] # Coût de frêt moyen pour 1 produit de la même commande


# Fonction pour récupérer la durée depuis la dernière commande pour chaque client:
def date_mini(x):
    if type(x) == pd.core.arrays.timedeltas.TimedeltaArray:
        return x.min()
    else:
        return x

# Application de la fonction au dataframe:
df_fe['Recency'] = df_fe['Recency'].apply(date_mini)

# Vérification sur les clients ayant commandé plus de 6 fois:
df_fe[df_fe['Orders'] > 6]
```
